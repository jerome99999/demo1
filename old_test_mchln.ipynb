```python
def rank_washing_machines(scores_df, prices_df):
    # Your code goes here
    new_df=(scores_df.groupby('BrandName')['Score'].mean()).reset_index()
    new_df=new_df.rename(columns={'Score':'AvgScore'})
    inner_merged_df = new_df.merge(
      prices_df,
      left_on = ["BrandName"],
      right_on = ["BrandName"],
      how = "inner"
    )
    inner_merged_df['prix_aj']=inner_merged_df['Price']/inner_merged_df['AvgScore']
    inner_merged_df=inner_merged_df.sort_values('prix_aj', ascending = True,ignore_index=True)
    inner_merged_df=inner_merged_df.drop(['prix_aj'], axis=1)
    columns_titles = ["BrandName","Price","AvgScore"]
    inner_merged_df=inner_merged_df.reindex(columns=columns_titles)
    return inner_merged_df
```

```python
def tokenize_sentence(sentence):
    punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    for ele in sentence:
        if ele in punc:
            sentence = sentence.replace(ele, "")
    new_sentence1=sentence.split()
    new_sentence2_set=set(new_sentence1)
    new_sentence2 = [e for e in new_sentence2_set]
    new_sentence2.sort() 
    answer=[new_sentence2.index(e) for e in new_sentence1]
    return answer
```

```python
import numpy as np
def standardize_data(arr):
    # Your code goes here
    centre_gravite = np.mean(arr, axis=1)
    row,_l=arr.shape
    #print(row,col)
    cg_r=centre_gravite.reshape(row,1)
    Y=arr-cg_r
    std_r=np.std(arr, 1).reshape(row,1)
    Z=Y/std_r
    # si ecart type =0 => mettre valeur np.inf a la place de nan
    Z[np.isnan(Z)] = np.inf
    return Z

```

```python
import pandas as pd
import numpy as np


def init_data():
    FirstName = pd.Series([ 'Jack','Sam','Jenn','Beth'])
    LastName = pd.Series(['Doe','Williams','Smith','Parker'])
    GradYear = pd.Series([2021,2022,2023,2022])
    MeanScore = pd.Series([ 90.6,88.7,74.2,89.5])
    Medianscore = pd.Series([90.0,88.5,75.5,89.1])
    df = pd.DataFrame({'FirstName': FirstName, 'LastName': LastName,'GradYear':GradYear,'MeanScore':MeanScore,'Medianscore':Medianscore })
    return df

def df2matrix(df):
    # Your code goes here
    row,col=df.shape
    array1 = df[df.columns[3:]].to_numpy()
    array2=np.zeros((row,col-2))
    array3 = np.column_stack((array1, array2))
    for i in range(len(df)):
        index_maj=int(str(df.loc[i,'GradYear'])[-1])
        array3[i,index_maj+1]=1
    return array3
```

```python
my_data=init_data()
print(my_data, '\n')   # Example input
print(df2matrix(my_data))
```
