```python
# version toutes colonnes
def handle_cols(df, cols):
    df = df.copy()
    for col in cols:
        Q1, Q3 = df[col].quantile([0.25, 0.75])
        IQR = Q3 - Q1
        lim_inf, lim_sup = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR

        min_no_col = df.loc[df[col] >= lim_inf, col].min()
        max_no_col = df.loc[df[col] <= lim_sup, col].max()

        df[col] = df[col].clip(lower=min_no_col, upper=max_no_col)

    return df
df = init_data()
df_clean = handle_cols(df, ['C1', 'C2', 'C3'])
#df_clean = handle_cols(df, df.columns)  # toutes les colonnes
df_clean
```

```python
# version tracking des valeurs remplacées
def handle_cols_verbose(df, cols):
    df = df.copy()
    outliers_report = {}

    for col in cols:
        Q1, Q3 = df[col].quantile([0.25, 0.75])
        IQR = Q3 - Q1
        lim_inf, lim_sup = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR

        # Repères de remplacement
        min_no_col = df.loc[df[col] >= lim_inf, col].min()
        max_no_col = df.loc[df[col] <= lim_sup, col].max()

        # Enregistrer les outliers
        outliers = df.loc[(df[col] < lim_inf) | (df[col] > lim_sup), col]
        outliers_report[col] = {
            'count': len(outliers),
            'values': outliers.values.tolist(),
            'lim_inf': lim_inf,
            'lim_sup': lim_sup,
            'replacement_min': min_no_col,
            'replacement_max': max_no_col
        }

        # Apaiser les excès
        df[col] = df[col].clip(lower=min_no_col, upper=max_no_col)

    return df, outliers_report
df = init_data()
df_clean, report = handle_cols_verbose(df, df.columns)

# Et pour lire le rapport :
from pprint import pprint
pprint(report)
```

```python
def rank_washing_machines(scores_df, prices_df):
    # Your code goes here
    # Calcule le score moyen (AvgScore) par marque.
    new_df=(scores_df.groupby('BrandName')['Score'].mean()).reset_index()
    # Fait une jointure interne avec le DataFrame des prix, pour ne garder que les marques communes aux deux tableaux.
    new_df=new_df.rename(columns={'Score':'AvgScore'})
    inner_merged_df = new_df.merge(
      prices_df,
      left_on = ["BrandName"],
      right_on = ["BrandName"],
      how = "inner"
    )
    #Calcule un prix ajusté = prix divisé par score moyen : plus il est petit, meilleur est le rapport qualité/prix.
    inner_merged_df['prix_aj']=inner_merged_df['Price']/inner_merged_df['AvgScore']
    #Trie du meilleur rapport qualité/prix au moins bon.
    inner_merged_df=inner_merged_df.sort_values('prix_aj', ascending = True,ignore_index=True)
    #Supprime la colonne intermédiaire une fois le tri effectué.
    inner_merged_df=inner_merged_df.drop(['prix_aj'], axis=1)
    #Réorganise les colonnes dans l’ordre souhaité.
    columns_titles = ["BrandName","Price","AvgScore"]
    inner_merged_df=inner_merged_df.reindex(columns=columns_titles)
    return inner_merged_df
```

```python
def rank_washing_machines(scores_df, prices_df): 
    # Calcul du score moyen par marque
    avg_scores = scores_df.groupby("BrandName", as_index=False)["Score"].mean()
    avg_scores.rename(columns={"Score": "AvgScore"}, inplace=True)

    # Jointure avec les prix
    merged_df = avg_scores.merge(prices_df, on="BrandName", how="inner")

    # Tri par rapport qualité/prix
    merged_df["Ratio"] = merged_df["Price"] / merged_df["AvgScore"]
    ranked_df = merged_df.sort_values("Ratio", ascending=True, ignore_index=True)

    # Nettoyage et sélection finale des colonnes
    return ranked_df[["BrandName", "Price", "AvgScore"]]
scores_df1, prices_df1=init_data3()
rank_washing_machines(scores_df1, prices_df1)
```

```python
def tokenize_sentence(sentence):
    punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    # supprimer la ponctuation
    for ele in sentence:
        if ele in punc:
            sentence = sentence.replace(ele, "")
    # separe les mots
    new_sentence1=sentence.split()
    new_sentence2_set=set(new_sentence1)
    new_sentence2 = [e for e in new_sentence2_set]
    # tri des mots
    new_sentence2.sort() 
    print(new_sentence2)
    # construit la liste des indices depuis la premiere phrase
    answer=[new_sentence2.index(e) for e in new_sentence1]
    return answer
```

```python
# version warning si std=0
import numpy as np
def standardize_data(arr):
    # Your code goes here
    # Calcule la moyenne de chaque ligne.axis=1 → on réduit les colonnes.
    centre_gravite = np.mean(arr, axis=1)
    #On remet la moyenne en colonne pour la soustraction.
    row,_l=arr.shape
    #print(row,col)
    cg_r=centre_gravite.reshape(row,1)
    
    # Centrage : soustrait la moyenne à chaque élément de la ligne
    Y=arr-cg_r
    # Divise chaque valeur par l’écart-type → standardisation.
    # On obtient des valeurs avec moyenne 0 et écart-type 1.
    std_r=np.std(arr, 1).reshape(row,1)
    # NumPy tente la division par zéro, ce qui produit déjà des NaN, 
    # et déclenche l'avertissement RuntimeWarning: invalid value encountered in divide.
    Z=Y/std_r
    # si ecart type =0 => mettre valeur np.inf a la place de nan
    # Si l’écart-type est 0 (ligne constante), on a une division par 0 → NaN.
    # remplacer ces NaN par inf.
    Z[np.isnan(Z)] = np.inf
    return Z

```

```python
# VERSION SANS WARNING DE DIV ZERO : on remplace 0 par 1 temporairement
import numpy as np

import numpy as np

import numpy as np

def standardize_data(arr):
    means = np.mean(arr, axis=1, keepdims=True)              # (n,1)
    stds = np.std(arr, axis=1, keepdims=True)                # (n,1)
    stds_safe = np.where(stds == 0, 1, stds)                 # (n,1)

    Z = (arr - means) / stds_safe                            # (n,m)

    # masque logique 2D (n,1) → on le répète pour correspondre à Z (n,m)
    mask = (stds == 0)
    print( mask)
    # Étendre le masque (broadcasting manquant)
    # mask est (n,1) → on le réplique horizontalement autant de fois qu’il y a de colonnes.
    # mask_full devient (n, m), comme Z.
    mask_full = np.repeat(mask, arr.shape[1], axis=1) 
    print(mask_full)# (n,m)
    # partout ou c'est True je remplace
    Z[mask_full] = np.inf

    return Z

arr = np.array([
    [1, 2, 3],      # std != 0
    [4, 4, 4],      # std == 0
    [10, 20, 30]    # std != 0
])
print(standardize_data(arr))
```

```python
#Version sans masque logique — avec np.where
import numpy as np

def standardize_data(arr):
    means = np.mean(arr, axis=1, keepdims=True)
    stds = np.std(arr, axis=1, keepdims=True)

    # Empêche la division par zéro : on remplace les zéros par 1 temporairement
    stds_safe = np.where(stds == 0, 1, stds)

    Z = (arr - means) / stds_safe

    # Là où std était 0 → on met np.inf
    Z = np.where(stds == 0, np.inf, Z)

    return Z
print(standardize_data(arr))
```

```python
# version boucle sur ligne et colonne a l'ancienne
import numpy as np

def standardize_data_lowtech(arr):
    # on veut un np.ndarray , on se premuni des listes de liste
    arr = np.array(arr)  # au cas où on reçoit une liste
    n_rows, n_cols = arr.shape
    Z = np.empty_like(arr, dtype=float)
    # boucle sur les lignes
    for i in range(n_rows):
        row = arr[i]
        mean = np.mean(row)
        std = np.std(row)

        if std == 0:
            Z[i] = np.full(n_cols, np.inf)  # tous les éléments deviennent inf
        else:
            Z[i] = (row - mean) / std

    return Z
arr = np.array([
    [1.2, -0.3, 1.3],
    [1.2, 2., 0.3]
]).astype(np.float64)

print(standardize_data(arr))
arr = np.array([[1, 2], [3, 4]])
print(standardize_data_lowtech(arr))
# cas probematique : ca passe qd meme
arr = [[1, 2], [3, 4]]
standardize_data_lowtech(arr)
```

```python
# version boucle sur ligne + last digit
def df2matrix_v3(df):
    row, col = df.shape
    
    # Récupérer les colonnes de score (ici colonnes 2 et 3 = MeanScore, MedianScore)
    score_data = df.iloc[:, 3:5].to_numpy()  # Colonnes index 3 et 4
    
    # Créer une matrice à zéros pour accueillir le codage (GradYear etc.)
    num_extra_cols = col - 3  # 1 colonne pour GradYear en plus des scores
    extra_data = np.zeros((row, num_extra_cols))
    
    # Fusion : scores + colonnes codées à 0
    matrix = np.column_stack((score_data, extra_data))
    
    # Encodage basé sur le dernier chiffre de GradYear
    for i in range(row):
        year = df.loc[i, 'GradYear']
        last_digit = int(str(year)[-1])  # ou year % 10
        if last_digit < num_extra_cols:
            matrix[i, last_digit + 2] = 1  # +2 pour décaler après 2 colonnes de score
    
    return matrix
my_data=init_data()
print(my_data, '\n')   # Example input
print(df2matrix_v3(my_data))
```

## df2matrix_manual : sans boucle ni get_dummies

Ultra-performant (tout en numpy vectorisé ).

Pas de boucle lente.

Plus de contrôle sur l’ordre des années.

Fonctionne même si les années ne sont pas triées dans df.

```python
def df2matrix_manual(df):
    # Étape 1 : extraction des scores
    #  extrait directement les deux colonnes numériques en numpy array.=> un tableau shape (n, 2)
    score_data = df[['MeanScore', 'MedianScore']].to_numpy()

    # Étape 2 : construction one-hot "à la main"
    #  extrait la colonne des années 
    years = df['GradYear'].to_numpy()
    # identifie les années uniques, triées. Pourquoi ?
    # Pour garantir un ordre stable des colonnes dans la matrice one-hot.
    # Le tri est important pour la cohérence.
    unique_years = np.sort(np.unique(years))  # [2021, 2022, 2023]
    
    #On convertit chaque année en indice :
    # 2021 → 0, 2022 → 1, etc.  np.searchsorted est très rapide car unique_years est trié.
    
    year_indices = np.searchsorted(unique_years, years)  # indices de chaque année
    # On crée une matrice binaire pleine de 0, avec :
    # une ligne par étudiant
    # une colonne par année unique
    one_hot = np.zeros((len(years), len(unique_years)))
    # calcul vectorisé ici 
    # np.arange(len(years)) donne les lignes
    # year_indices donne les colonnes où mettre les 1
    # On remplit la matrice en une seule ligne (sans boucle !)
    one_hot[np.arange(len(years)), year_indices] = 1

    # Étape 3 : concaténation finale
    # hstack assemble les colonnes.
    return np.hstack([score_data, one_hot])

my_data=init_data()
print(my_data, '\n')   # Example input
print(df2matrix_manual(my_data))
```

## version Preprocess avec sklearn

Extraire les colonnes pertinentes : Nous sélectionnons uniquement les colonnes mood, restrictions et integrations.

Initialisation de OneHotEncoder :

Le paramètre drop='first' permet de supprimer la première colonne de chaque catégorie, ce qui évite la colinéarité (c'est-à-dire, éviter que la somme des colonnes ne soit égale à 1, ce qui est souvent une bonne pratique pour les modèles statistiques).

Le paramètre sparse=False permet de retourner un tableau numpy plutôt qu'une matrice creuse (spare matrix), ce qui est plus facile à manipuler dans ce cas.

Encodage : Le OneHotEncoder est appliqué sur les colonnes catégorielles mood, restrictions, et integrations. Cela transforme chaque catégorie en une colonne binaire où la valeur est 1 si la catégorie est présente et 0 sinon.

Création du DataFrame final : Nous créons un DataFrame avec les résultats de l'encodage et générons des noms de colonnes en utilisant get_feature_names_out(), ce qui nous donne des noms comme mood_crash, restrictions_banned, etc.

Retour : La fonction retourne le DataFrame encodé.



```python
# version avec sklearn.preprocessing import OneHotEncoder
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import numpy as np

def preprocessX(df):
    # Extraire les colonnes mood, restrictions, et integrations
    selection_data = df[['mood', 'restrictions', 'integrations']]

    # Initialiser l'encodeur OneHotEncoder
    encoder = OneHotEncoder(drop='first', sparse_output=False)

    # Encoder les colonnes catégorielles
    encoded_data = encoder.fit_transform(selection_data)

    # Créer un DataFrame avec les résultats de l'encodage et les noms des colonnes
    encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(selection_data.columns))

    # Retourner le DataFrame encodé
    return encoded_df

preprocessX(dataframe)
```

```python
# Exemple d'appel à la fonction avec autre data
data = pd.DataFrame({
    'mood': ['crash', 'stable', 'good', None, 'Stable'],
    'restrictions': ['banned', 'Regulated', 'free', 'regulated', 'free'],
    'integrations': ['noIntegration', 'PayPal', 'nointegration', 'yandex', 'Paypal'],
    'media': [0.6, 0.7, 0.8, 0.9, 0.9],
    'price': [9008, 16065, 17076, 19341, 4789]
})

processed_data = preprocessX(data)
print(processed_data)
```
