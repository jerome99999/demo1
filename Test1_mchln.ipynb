```python

def cluster_points(points, centers):
    # Initialise un dictionnaire pour stocker les indices de points pour chaque centre
    clusters = {center_name: [] for center_name in centers}
    
    # Pour chaque point, déterminer son centre le plus proche
    for idx, point in enumerate(points):
        distances = {}
        for center_name, center_coord in centers.items():
            # Conversion du centre en array numpy pour le calcul
            center_coord = np.array(center_coord)
            # Calcul de la distance euclidienne
            distances[center_name] = np.linalg.norm(point - center_coord)
        
        # Trouver le centre avec la plus petite distance
        closest_center = min(distances, key=distances.get)
        # Ajouter l'indice du point à la liste du centre correspondant
        clusters[closest_center].append(idx)
    
    return clusters

# Exemple d'utilisation
points, centers = init_exemple()
resultat = cluster_points(points, centers)
print(resultat)
```

```python
# Fonction pour calculer la distance euclidienne entre deux points 3D
def distance(p, q):
    return ((p[0] - q[0])**2 + (p[1] - q[1])**2 + (p[2] - q[2])**2) ** 0.5

def cluster_points(points, centers):
    clusters = {name: [] for name in centers}
    # Pour chaque point, on calcule la distance entre le point et chaque centre
    for i, point in enumerate(points):
        distances = {}
        for name, center in centers.items():
            distances[name] = distance(point, center)
        # On détermine le centre le plus proche
        closest = min(distances, key=distances.get)
        clusters[closest].append(i)
    return clusters

# Exemple d'utilisation
points, centers = init_exemple()
print(cluster_points(points, centers))
```

### calcul vectorize

```python
def cluster_points_vectorized(points, centers):
    # Conversion du dictionnaire des centres en un tableau numpy et liste de noms
    center_names = list(centers.keys())
    center_coords = np.array([centers[name] for name in center_names])
    
    # Calcul vectorisé des distances euclidiennes :
    # points[:, None, :] crée un tableau de forme (N, 1, 3)
    # center_coords[None, :, :] crée un tableau de forme (1, M, 3)
    # La soustraction se fait en broadcast pour obtenir un tableau (N, M, 3)
    # On calcule ensuite la somme des carrés sur l'axe 2 et la racine carrée
    distances = np.sqrt(((points[:, None, :] - center_coords[None, :, :])**2).sum(axis=2))
    
    # Pour chaque point, on récupère l'indice du centre le plus proche
    indices_min = np.argmin(distances, axis=1)
    
    # Reconstruction du dictionnaire de clusters avec indices des points
    clusters = {name: [] for name in center_names}
    for point_idx, center_idx in enumerate(indices_min):
        clusters[center_names[center_idx]].append(point_idx)
    
    return clusters

points, centers = init_exemple()
result = cluster_points_vectorized(points, centers)
print(result)
```

```python
# visu avec distance et legende
def plot_clusters(points, centers, clusters):
    # Création du graphique 3D
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    # Liste de couleurs pour différencier les clusters
    colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k']
    
    # Tracé des points appartenant à chaque cluster
    print(clusters.items())
    for idx, (center_name, point_indices) in enumerate(clusters.items()):
        color = colors[idx % len(colors)]
        pts = points[point_indices]  # Extraction des points du cluster
        ax.scatter(pts[:, 0], pts[:, 1], pts[:, 2],
                   c=color, marker='o', s=60, label=f'point {center_name} cluster')
        
        # Récupération des coordonnées du centre associé
        center = centers[center_name]
        
        # Tracer pour chaque point une ligne pointillée vers le centre et afficher la distance sur la ligne
        for p in pts:
            ax.plot([p[0], center[0]], [p[1], center[1]], [p[2], center[2]], 'k--')
            distance_val = np.linalg.norm(np.array(p) - np.array(center))
            ax.text((p[0] + center[0]) / 2,
                    (p[1] + center[1]) / 2,
                    (p[2] + center[2]) / 2,
                    f"{distance_val:.2f}", color='r', fontsize=10)
    
    # Tracé des centres avec un marqueur distinctif (ici un 'x')
    center_names = list(centers.keys())
    for idx, center_name in enumerate(center_names):
        center_coord = centers[center_name]
        color = colors[idx % len(colors)]
        ax.scatter(center_coord[0], center_coord[1], center_coord[2],
                   c=color, marker='x', s=200, linewidths=3,
                   label=f'{center_name} center')
        
    ax.plot([p[0], center[0]], [p[1], center[1]], [p[2], center[2]], 'k--'); 
    ax.text((p[0]+center[0])/2, (p[1]+center[1])/2, (p[2]+center[2])/2, f"{np.linalg.norm(np.array(p)-np.array(center)):.2f}", color='r', fontsize=10)

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title("Regroupement de points en clusters")
    ax.legend(loc='best')
    for i, p in enumerate(points): ax.text(p[0], p[1], p[2], str(i), fontsize=12, color='black')
    plt.show()
```
