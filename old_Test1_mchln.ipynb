```python

def init_data0():
    random.seed(10)
    my_clust=np.array([])
    for j in range(10):
        list_coord=[round(random.uniform(-100, 100),2) for x in range(3)]
        my_clust=np.append(my_clust,[np.array(list_coord)])
    my_clust=my_clust.reshape(10,3)
    my_center={}
    for k in range(3):
        name_centre='Center'+str(k+1)
        my_center[name_centre]=[round(random.uniform(-100, 100),2) for x in range(3)]
    return (my_clust,my_center)
```

```python
A=np.array([[3,6],[2,5],[7,3],[1,5]])
B=np.array([[1,2],[3,6],[4,9]])
cg=np.array([1,2,3])
def add_edm(A,B):
    p1 = np.sum(A**2, axis=1)[:,np.newaxis]
    p2 = np.sum(B**2, axis=1)
    p3 = -2 * np.dot(A,B.T)
    return np.round(np.sqrt(p1+p2+p3),2)
res=add_edm(A,B)
print(res)
 
def distance_eucli(v1, v2):
    return np.sqrt(np.sum((v1 - v2) ** 2)) 
print("ditance entre (3,6) et (1,2)")
res2=distance_eucli(A[0], B[0])
print(res2)
res2=distance_eucli(A[0], B[1])
print(res2)
res2=distance_eucli(A[0], B[2])
print(res2)
```

```python
# Python code to find Euclidean distance
# using dot()

import numpy as np

# initializing points in
# numpy arrays
point1 = np.array((1, 2, 3))
point2 = np.array((1, 1, 1))

# subtracting vector
temp = point1 - point2

# doing dot product
# for finding
# sum of the squares
sum_sq = np.dot(temp.T, temp)

# Doing squareroot and
# printing Euclidean distance
print(np.sqrt(sum_sq))
```
